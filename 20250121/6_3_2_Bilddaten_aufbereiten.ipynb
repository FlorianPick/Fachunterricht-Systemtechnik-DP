```python
from skimage.io import imread, imshow
image = imread("schnecke.jpg", as_gray = True)
imshow(image)
```

```python
#type(image)
image.shape
```

```python
cropped = image[250:750, 190:600]
imshow(cropped)
```

```python
from skimage import exposure
contrast = exposure.adjust_log(cropped)
imshow(contrast)
```

```python
from skimage import feature
edges = feature.canny(contrast, sigma=3)
imshow(1 - edges)
#imshow(edges)
```

```python
from skimage.segmentation import chan_vese
cv = chan_vese(contrast)
imshow(cv)
```

```python
from skimage.transform import resize
resized = resize(contrast, (40, 40))
imshow(resized)
```

```python
resized2 = resize(contrast, (20, 20))
imshow(resized2)
```

```python
resized3 = resize(edges, (20, 20))
imshow(resized3)
```

```python
resized4 = resize(cv, (20, 20))
imshow(resized4)
```
