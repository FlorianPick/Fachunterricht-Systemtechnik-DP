## Eine minimale Anwendung (OO-Stil):

````python
import tkinter as tk                    # tkinter importieren, umbenannt zu tk
class Application(tk.Frame):            # Klasse "erbt" von "Frame" Klasse
    def __init__(self, master=None):
        tk.Frame.__init__(self, master) # Init Methode der Elternklasse "Frame"
        self.pack()                     # Erforderlich zur Darstellung
        #self.createWidgets()            # Aufruf der Widget-Erzeugung
    def createWidgets(self):
        self.quitButton = tk.Button(self, text='Quit', command=self.master.destroy) # erzeugt "Quit" button
        self.quitButton.pack()          # Darstellung
app = Application()                     # Hauptprogramm startet durch Instantiierung
app.master.title('Hallo, tkinter')      # Titel erzeugen
app.createWidgets()
app.mainloop()                          # Event Loop starten (Maus- und Keyboard)
app.
````

## Geometriemanager
Ein Geometriemanager verhandelt die Positionierung der Widgets.  
Es existieren 3 verschiedene Geometriemanager in tkinter.  
**Dies sind:**
- Pack (Alle Widgets vertikal übereinander gestapelt, side=Left/right kontrolliert die Ausrichtung)
- Place (explizite x- und y-Positionsangabe, unflexibel und kompliziert)
- Grid (für eine Tabellenähnliche Plazierung mittels Zeilen und Spalten; für kompliziertere Oberflächen, row/column/columnspan/rowspan/sticky options)

Der Geometriemanager .pack() eignet sich, um Widgets in einer Spalte übereinander anzuordnen: 

````python
import tkinter as tk
def druck():
    print(ent.get())
    
root = tk.Tk()
label = tk.Label(root, text="Name")
label.pack()                        # .pack () in der Zeile des Erstellens ist meist problematisch
ent = tk.Entry(root)
ent.pack(fill = tk.X,expand = 1)                          
#ent.pack(fill = tk.NONE ,expand = 1)                          
but1 = tk.Button(root,text="Accept",command = druck).pack(fill = tk.BOTH,expand = 1) # besser ist eine neue Zeile mit label.pack() und ent.pack()
but = tk.Button(root,text="Quit",command = root.destroy).pack(fill = tk.BOTH,expand = 1)
root.mainloop( )
````

## Optionen des .pack() Geometriemanagers



Zur beidseitigen Ausdehnung bis zur Breite des übergeordneten Fensters fill = tk.X verwenden.

Der Geometriemanager .grid() platziert Widgets anhand von Zeilen und Spalten, ähnlich eines Excel-Sheets. Die Plazierung erfolgt über eine Zeilen- und Spaltenangeabe: 

````python
import tkinter as tk

root = tk.Tk()

def readentry():
    v1 = ent1.get()
    print(f'Vorname : {v1}')

l1 = tk.Label(root, text="First Name")
l1.grid(row=0)  # Zeile 0
l2 = tk.Label(root, text="Last Name")
l2.grid(row=5)  # Zeile 1

ent1 = tk.Entry(root)
ent1.grid(row=0, column=1)  # Zeile 0, Spalte 1

ent2 = tk.Entry(root)
ent2.grid(row=5, column=1)  # Zeile 1, Spalte 1

ent3 = tk.Entry(root)
ent3.grid(row=21, column=2)  # Zeile 1, Spalte 2

ent4 = tk.Entry(root)
ent4.grid(row=20, column=2)  # Zeile 1, Spalte 2

b1 = tk.Button(root, text="Show", command=readentry)
b1.grid(row=10, column = 0, columnspan=3, sticky=tk.E + tk.W)

bq = tk.Button(root, text="Press Me to Quit", command=root.destroy)
bq.grid(row=20, column = 0, columnspan=2, rowspan=2, sticky=tk.E + tk.W + tk.S + tk.N)
root.mainloop()
````

Für User-interaktion werden tkinter events (Ereignisse) mit „bind“ zu Reaktionen verbunden, i.a. dem Aufruf einer Funktion oder Methode (Ausnahme: Widget „Button“ ist standardgemäß mit dem <Button-1> Event verbunden). Die Verbindung erfolgt mit dem “command=“ Eintrag.  
Es existieren u.a. die folgenden Events: 

    
https://python-course.eu/tkinter/events-and-binds-in-tkinter.php

Um mit den Widgets Variablen auszutauschen, existieren die folgenden tkinter Variablentypen:  
StringVar(), IntVar(), DoubleVar() und BooleanVar().  
Die Variablen werden bei der Erstellung der Widgets als variable/textvariable angegeben.  
Die enthaltenen Werte werden mit den Methoden .get() und .set() aus den Variablen gelesen und geschrieben.  
```
import tkinter as tk
d = tk.DoubleVar() # Enthält float; default Wert: 0.0
i = tk.IntVar() # Enthält int; default Wert: 0
s = tk.StringVar() # Enthält string; default Wert ''
Entry widget: textvariable = s, s.get()/s.set()
Radiobutton widget: variable = s, s.get()/s.set()
Scale widget: variable, from_, to = d,i oder s; s.set()/get() 
```

## Beispiel für die Verwendung von tkinter-Variablen:

````python
import tkinter as tk

def show_entry():  # Verwendung von Tkinter-Variablen
    print(f"You typed: {s.get()}")  # Variable auslesen
    s.set("nächste Eingabe")  # Variable schreiben

root = tk.Tk()  # Dieser Befehl muss am Anfang stehen
s = tk.StringVar()  # Dieser Befehl darf nicht vor tk.Tk() stehen
label = tk.Label(root, text="Enter Text:");
label.pack()
ent1 = tk.Entry(root, textvariable=s)
ent1.pack()
s.set("mike")  # Variable schreiben
but1 = tk.Button(root, text='Show', command=show_entry);
but1.pack(fill=tk.X, expand=1)
but2 = tk.Button(root, text='Quit', command=root.destroy)  # fuer spyder: root.master.destroy
but2.pack(fill=tk.BOTH, expand=1)
root.mainloop()
````

Manchmal besteht die Möglichkeit, Werte direkt aus Widgets zu lesen und zu schreiben:

````python
import tkinter as tk
def show_entry():
    print("You typed: %s" % (ent1.get(),)) # Inhalt auslesen ohne Variable
    ent1.delete(0,tk.END)                  # Inhalt loeschen tk.END --> letztes Zeichen
root = tk.Tk()                             # Muss am Anfang stehen
label = tk.Label(root, text="Enter Text:"); label.pack()
ent1 = tk.Entry(root); ent1.pack()         # Entry ohne textvariable
ent1.insert(tk.END,"mike")                 # Entry Wert einfuegen
#ent1.insert(2,"mike")                # Entry Wert einfuegen (Test)
but1 = tk.Button(root, text='Show', command=show_entry);
but1.pack(fill=tk.X,expand=1)
but2 = tk.Button(root, text='Quit', command=root.destroy)
but2.pack(fill=tk.BOTH,expand=1)
root.mainloop()
````

## Beispiel: Scale Widget mit DoubleVar

````python
# Python program to demonstrate
# scale widget
  
import tkinter as tk
root = tk.Tk()  
root.geometry("400x150") 
root.title("Scale Widget")
v1 = tk.DoubleVar()
  
def show1():  
    sel = "Horizontal Scale Value = " + str(v1.get())
    l1.config(text = sel, font = ("Courier", 14))  
    b1.config(text = 'Next Value')
    
s1 = tk.Scale( root, variable = v1, 
           from_ = 1, to = 100, orient = "horizontal")
  
l3 = tk.Label(root, text = "Horizontal Scaler")
  
b1 = tk.Button(root, text ="Display Horizontal", 
            command = show1, 
            bg = "yellow")  
b2 = tk.Button(root, text="close", command=root.destroy)

l1 = tk.Label(root)
  
s1.pack(anchor = CENTER) 
l3.pack()
b1.pack(anchor = CENTER)
b2.pack(anchor = W)
l1.pack() 

root.mainloop()
````

````python
### Beispiel mit place-Layoutmanager ###
import tkinter as tk

root = tk.Tk()
root.title("Place layout Example")
root.geometry("500x500+50+100")

def display_selection(event):
    selection = cities_listbox.curselection()
    print(cities_listbox.get(selection))

# Label to display the question
tk.Label(
    root,
    text="Which of the following cities would you like to travel to?",
    wraplength=200,
).place(x=50, y=20)

# Listbox to display the cities
cities_listbox = tk.Listbox(root, selectmode=tk.BROWSE, width=24)
cities_listbox.place(x=40, y=65)
cities = ["Beijing", "Singapore", "Tokyo", "Dubai", "New York"]
for city in cities:
    cities_listbox.insert(tk.END, city)

# Bind the listbox's selection
cities_listbox.bind("<<ListboxSelect>>", display_selection)
# Button to close the app
end_button = tk.Button(root, text="End", command=root.destroy)
end_button.place(x=125, y=250)

root.mainloop()
````
