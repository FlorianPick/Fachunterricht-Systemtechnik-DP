# Verteilung Zufallsgenerator
- Lassen Sie sich, per Zufallsgenerator, eine Zufallszahl zwischen 1 und 6 generieren.
- Definieren Sie eine Variable, die die Anzahl der Durchläufe bestimmt.
- Bei jedem Durchlauf wird eine neue Zufallszahl generiert und abhängig davon ein Zähler für diese Zahl um 1 erhöht.
Nach Beendigung der Durchläufe sollen folgende Werte pro Zähler (1 bis 6) berechnet und ausgegeben werden:
-- Zählerstand
-- Abweichung vom Wert (Durchläufe / 6) als Zahl und in %

**Beispiel für 1.000.000 Durchläufe**
```python
Durchläufe  :	1000000
Durchschnitt:	166666.67
Zähler 1: 166799	Diff:  132.33 	%: 0.08
Zähler 2: 166675	Diff:    8.33 	%: 0.01
Zähler 3: 167345	Diff:  678.33 	%: 0.41
Zähler 4: 166866	Diff:  199.33 	%: 0.12
Zähler 5: 166073	Diff: -593.67 	%: 0.36
Zähler 6: 166243	Diff: -423.67 	%: 0.25
```

**Führen Sie das Programm mit 100, 1.000, 10.000, 100.000 und 1.000.000 Durchläufen aus und erläutern Sie die Ergebnisse.**

````python
from random import randint
from time import time_ns
###################
### Initialisierung
###################
c1 = c2 = c3 = c4 = c5 = c6 = 0         # Zähler für Zahlen 1 bis 6
#
z = 0
d = 1000000        # Anzahl der Durchläufe
d6 = d/6           # 1/6 der Durchläufe
start = time_ns()
###################
### Berechnung
###################
while z < d:
    z += 1
    x = randint(1,6)         # Zufallszahl zwischen 1 und 6
    if x == 1: c1 += 1       # Zähler 1 + 1
    elif x == 2: c2 += 1     # Zähler 2 + 1
    elif x == 3: c3 += 1     # Zähler 3 + 1
    elif x == 4: c4 += 1     # Zähler 4 + 1
    elif x == 5: c5 += 1     # Zähler 5 + 1
    elif x == 6: c6 += 1     # Zähler 6 + 1
########################
# Ausgabe der Ergebnisse
########################
runtime = (time_ns() - start) /1e9
print(f'Laufzeit    :\t{runtime:.4f} s')
print(f'Durchläufe  :\t{d}')    
print(f'Durchschnitt:\t{d6:.2f}')    
print(f'Zähler 1: {c1}\tDiff: {c1 - d6:>8.2f} \t: {abs(c1-d6) / d6 *100:>6.2f} %') 
print(f'Zähler 2: {c2}\tDiff: {c2 - d6:>8.2f} \t: {abs(c2-d6) / d6 *100:>6.2f} %') 
print(f'Zähler 3: {c3}\tDiff: {c3 - d6:>8.2f} \t: {abs(c3-d6) / d6 *100:>6.2f} %') 
print(f'Zähler 4: {c4}\tDiff: {c4 - d6:>8.2f} \t: {abs(c4-d6) / d6 *100:>6.2f} %') 
print(f'Zähler 5: {c5}\tDiff: {c5 - d6:>8.2f} \t: {abs(c5-d6) / d6 *100:>6.2f} %') 
print(f'Zähler 6: {c6}\tDiff: {c6 - d6:>8.2f} \t: {abs(c6-d6) / d6 *100:>6.2f} %')
print()
print(f'Summe   : {(c1 + c2 + c3 + c4 + c5 + c6)}')
````

````python
# Variante mit Liste für Zähler
# Als Funktion
from random import randint
from time import time_ns
def wuerfeln(d):
    ###################
    ### Initialisierung
    ###################
    #zaehler = [0,0,0,0,0,0]       # Zähler Zähler 1 bis 6
    zaehler = [0]*6               # Zähler Zähler 1 bis 6
    dabw = 0                      # Variable Durchschnitt Abweichungen absolut   
    #print(zaehler)
    #d = 1000000                   # Anzahl der Durchläufe
    durchschnitt = d/6            # 1/6 der Durchläufe
    start = time_ns()
    ###################
    ### Berechnung
    ###################
    for i in range(d):
        x = randint(1,6)   # Zufallszahl zwischen 1 und 6
        zaehler[x-1] += 1  # Zähler x erhöhen
    ########################
    # Ausgabe der Ergebnisse
    ########################
    runtime = (time_ns() - start) /1e9
    print(f'Laufzeit    :\t{runtime:.4f} s')
    print(f'Durchläufe  :\t{d}')    
    print(f'Durchschnitt:\t{durchschnitt:.2f}')
    #print(list(enumerate(zaehler)))
    for n,i in enumerate(zaehler):
        abw = abs(i-durchschnitt) / durchschnitt *100  # Abweichung absolut
        dabw += abw                                    # Variable für Durchschnitt Abweichungen
        print(f'Zähler {n+1} : {i:>8}   Diff: {i - durchschnitt:>8.2f} \t: {abw:>6.2f} %') 
    print()
    print(f'Summe   : {sum(zaehler)}')
    return zaehler, dabw

#######################
### M A I N ###########
#######################
li = [100, 1000, 10000, 100000, 1000000, 10000000]      # Was soll berechnet werden?
for i in li:
    zx, dabw = wuerfeln(i)
    print(f'Summe der Abweichungen (absolut) : {dabw:5.2f} %')
    print('_'*50)
````
