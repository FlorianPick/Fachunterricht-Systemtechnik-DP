# Konditionen

sind Aussagen, die uns helfen zu prüfen, ob eine Bedingung erfüllt ist (oder nicht) und dann **eine Entscheidung zu treffen**:

Angenommen, wir wollen ein Programm für einen Nachtclub schreiben, der entscheidet, Leute hereinzulassen, **wenn** sie älter als 18 sind:

````python
# Frage nach dem Alter
#age = int(input('Wie alt sind Sie? '))
age = input('Wie alt sind Sie? ')
#print(type(age))
age = int(age)
#print(type(age))
# Prüfe ob die Person älter als 18 ist:
if age > 18:
    print('Willkommen im Club') # Einrückung 4 Leerzeichen; Block
    print('Komm herein')        # Block   
print(f'Alter ist {age}.')
````

_Einrückungen_ in if-Anweisungen sind sehr wichtig, da Python alle eingerückten Befehle als einen Codeblock versteht, der ausgeführt werden muss, wenn die Bedingung erfüllt ist.

age > 18 # Alter = 19, daher ist die Antwort **wahr**: Es bedeutet, dass die Bedingung erfüllt ist

Bedingungen werden auch als **Kontrollstrukturen** bezeichnet, da sie Entscheidungen treffen und die Ergebnisse abhängig von Bedingungen ändern können.

````python
# Frage nach dem Alter
age = int(input('Wie alt sind Sie? '))

# Prüfe ob die Person älter als 18 ist:
if age > 18:                    # Block einleiten mit :
    print('Willkommen im Club') # Einrückung 4 Leerzeichen
else:                           # Wenn keine Bedingung erfüllt wird 
    print('Leider sind Sie nicht alt genug!')
print(f'Alter ist {age}.')
````

Was ist, wenn jemand genau 18 ist?

Wir können Vergleichsoperatoren mit weiteren logischen Operatoren **and** / **or** kombinieren

````python
# Frage nach dem Alter
age = int(input('Wie alt sind Sie? '))

# Prüfe ob die Person älter als 18 ist:
if age > 18 or age == 18:      # Block einleiten mit :
    print('Willkommen im Club') # Einrückung 4 Leerzeichen
else: # Wenn keine Bedinguung erfüllt wird 
    print('Leider sind Sie nicht alt genug!')
````

## logischen Operatoren:

a **oder** b: ist wahr, wenn mindestens eine der Aussagen wahr ist

a **und** b: ist wahr, nur wenn beide Aussagen wahr sind



````python
age = 18
age > 18 # age = 18 also 18 > 18 is False
````

````python
age == 18 # age = 18 also 18 == 18
````

````python
# Wahrheitswerte Oder-Verknüpfung
print(False or False)
print(False or True)
print(True or False)
print(True or True)
````

````python
# Wahrheitswerte Und-Verknüpfung
print(False and False)
print(False and True)
print(True and False)
print(True and True)
````

Wir können die beiden Vergleichsoperatoren > und == auch wie folgt kombinieren:

````python
# Frage nach dem Alter
age = int(input('Wie alt sind Sie? '))

# Prüfe ob die Person älter als 18 ist:
if age >= 18:                   # Alternative mit anderem Operator
    print('Willkommen im Club') # Einrückung 4 Leerzeichen
else: # Wenn keine Bedinguung erfüllt wird 
    print('Leider sind Sie nicht alt genug!')
````

Auf die gleiche Weise werden < and == kombiniert:

````python
# Frage nach dem Alter
age = int(input('Wie alt sind Sie? '))

if age <= 17: # if age < 17 or age == 17:
if age < 18:  # Alternative; Vorsicht, Grenzen (Wertebereiche) genau bewerten!
    print('Leider sind Sie nicht alt genug!')
else: # Wenn keine Bedingung erfüllt wird 
    print('Willkommen im Club') # Einrückung 4 Leerzeichen
````

## Neu ab Python 3.10
**PEP 634: Structural Pattern Matching**  
*Structural pattern matching has been added in the form of a match statement and case statements of patterns with associated actions. Patterns consist of sequences, mappings, primitive data types as well as class instances. Pattern matching enables programs to extract information from complex data types, branch on the structure of data, and apply specific actions based on different forms of data.*

````python
status = 404
match status:
        case 400:
            print( "Bad request")
        case 404:
            print( "Not found")
        case 418:
            print( "I'm a teapot")
        case 401 | 403 | 404:
            print( 'Not allowed')
        case _:
            print( "Something else")
````
