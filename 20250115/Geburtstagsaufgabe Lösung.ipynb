Leiten Sie sich eine Berechnungsformel her und dokumentieren diese hier:  

Wahrscheinlichkeit für **50** Spieler mit mind. einem gleichen Geburtstag? (kein Schaltjahr!!)

Dies ist gleichbedeutend mit: nicht alle **50** Spieler haben an jeweils unterschiedlichen Tagen Geburtstag.

Das Gegenereignis entspricht (bei **365** möglichen Geburtsdaten) also $\binom{365}{50}$ Möglichkeiten.

Anzahl aller möglichen Geburtstagskombinationen bei 50 Spielern und 365 möglichen Geburtstagen:
$\frac{365^{50}}{50!}$ (Teilung durch **50!** um die Möglichkeiten der Vertauschung der gewählten Daten untereinander zu berücksichtigen - dies ist oben im Binomialkoeffizienten bereits berücksichtigt).

Die Wahrscheinlichkeit, dass alle Geburtsdaten unterschiedlich sind, entspricht dem Verhältnis der Anzahl dieser Kombinationsmöglichkeiten, also für das Gegenereignis:
  
   
$$ q = \frac{\binom{365}{50}}{\frac{365^{50}}{50!}} 
= \frac{\frac{365!}{50! \cdot (365-50)!}}{\frac{365^{50}}{50!}} 
= \frac{\frac{365!}{315!}}{365^{50}} 
= \frac{\prod_{n=316}^{365} n}{365^{50}}
= \frac{316 \cdot \cdots \cdot 365}{365^{50}}
= \prod_{n=316}^{365} \frac{n}{365} $$

```python
# Binomialkoeffizient (Kombinationsmöglichkeiten) 50 aus 365
# Alle 50 Studenten*innen haben an einem anderen Tag Geburtstag 
import math
x = math.comb(365, 50)
str = f'{x:,}'.replace(',', '.')
print(str)
```

```python
# Alle möglichen Geburtstagskombinationen bei 50 Studenten und 365 möglichen Geburtstagen
alle = 365**50/math.factorial(50)
str = f'{alle:,}'.replace(',', '.')
print(str)
```

```python
# Schreibe Sie Ihren Programmcode hier
#TIP: math.comb; math.factorial kann helfen!
import math
anz = 50
q = (math.comb(365,anz) * math.factorial(anz) / 365**anz)*100
print(f'Wahrscheinlichkeit für keinen gleichen Geburtstag: {q:2.6f} %')
# Die gesuchte Wahrscheinlichkeit ist also 1-q:
print(f'Wahrscheinlichkeit für mindestens 2 gleiche Geburtstage: {100 - q:2.6f} %')
```

$$ P(A_n) = \frac{\frac{365!}{315!}}{365^{50}}
