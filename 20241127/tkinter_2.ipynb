## tkinter Pop-Up Dialoge
- tkinter.Toplevel(): Container für andere Widgets: hier am Beispiel „Message“, „Button“


```
top_w = Toplevel()
msg = Label(top_w, text="Text")
msg.pack(fill="x")
button = Button(top_w, text=" Dismiss ",command=top_w.quit)
button.pack(fill="x")
top_w.mainloop()
```



````python
import tkinter as tk
top_w = tk.Toplevel()
msg = tk.Label(top_w, text="Text")
msg.pack(fill="x")
button = tk.Button(top_w, text=" Dismiss ",command=top_w.destroy)
button.pack(fill="x")
top_w.mainloop()
````

## Beispiel für Toplevel()
```
import tkinter as tk
#root application, can only have one of these.
root = tk.Tk()
#put a label in the root to identify the window.
label1 = tk.Label(root, text="""this is root
closing this window will shut down app""")
label1.pack()
#you can make as many Toplevels as you like
extra_window = tk.Toplevel(root)
label2 = tk.Label(extra_window, text="""this is extra_window
closing this will not affect root""")
label2.pack()
root.mainloop()
```

````python
import tkinter as tk
#root application, can only have one of these.
root = tk.Tk()
#put a label in the root to identify the window.
label1 = tk.Label(root, text="""this is root
closing this window will shut down app""")
label1.pack()
#you can make as many Toplevels as you like
extra_window = tk.Toplevel(root)
label2 = tk.Label(extra_window, text="""this is extra_window
closing this will not affect root""")
label2.pack()
root.mainloop()
````
