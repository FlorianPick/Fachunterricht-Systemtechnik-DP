<h2><font color='darkblue'>1. Warum müssen Daten visualisiert werden?</font></h2>

Daten können vom menschlichen Auge besser erfasst werden, wenn diese visuell dargestellt sind.

Weiterhin kann mit der Visualisierung ein Thema schneller erfasst werden, als wenn die Ergebnisse in einem Fließtext/Zahlen dargestellt sind. 

Du kannst quasi auf einem Blick das Thema und den Inhalt des Datensatzes erfassen.

Zudem werden mit der Visualisierung Beziehungen zwischen den Daten "sichtbar".

Nachfolgender Datensatz der die Passagierzahlen einer Airline zeigt, lässt sich visuell klarer erfassen.





Mit der Heatmap lässt sich schnell erkennen, dass die Passagierzahlen über die Jahre stetig steigen und insbesondere in den sommermonaten die Auslastung hoch ist.

```python
import seaborn as sns

df = sns.load_dataset('flights')

flights = df.pivot("month", "year", "passengers")

ax = sns.heatmap(flights,linewidths=.5,center=flights.loc["Jan", 1955])
```

<h2><font color='darkblue'>2. Was ist bei der Visualisierung zu beachten?</font></h2>

Auch bei der Visualisierung gilt die Frage zu klären "<span class="girk">Was will ich darstellen?</span>".

Danach richtet sich z.B. die Auswahl des Plots, die Farbgestaltung, die Ausrichtung/Hervorhebung einzelner Elemente etc.

Weiterhin kann es sein, dass du deine Daten weiter aufbereiten musst.
Stell dir vor du hast einen Datensatz mit Ländern. Du möchtest zeigen, wie oft etwas in den Ländern vorkommt. Wenn du nun in einer Chorophletenmap (Darstellung per Karte) die Häufigkeiten in jedem Land darstellst, kann es sein, dass die Farbverläufe nicht richtig zu erkennen sind. In diesem Fall ist es sinnvoll nicht auf Länderebene sondern auf Ebene der Kontinente zu gehen. Dann müsstest du die Länder erst den Kontinenten zuordnen. Mit dieser reduzierten Darstellung sind die Farbstärken, die die Häufigkeiten widerspiegeln deutlicher zu erkennen.

Merke: Wenn Farbverläufe nicht richtig zu erkennnen sind, ist es empfehlenswert die Daten weiter zu gruppieren.

Gleiches gilt auch mit Legenden. Wenn die Legende zu lang ist, weil zu viele Elemente enthalten sind, ist dies ein Zeichen dafür, dass du die Themen zusammenfassen solltest.

Wenn du einzelne Besonderheiten darstellen möchtest, kannst du z.B. in einem zweiten Bild auf diese Besonderheiten eingehen.
